```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
with open(r'C:\Users\Gaurav\OptimAlg\AlternateAdam\train_logs_og.txt', 'r') as file:
    data1 = [float(line.strip()) for line in file]

with open(r'C:\Users\Gaurav\OptimAlg\AlternateAdam\train_logs_mod.txt', 'r') as file:
    data2 = [float(line.strip()) for line in file]
```

```python
plt.plot(data1, label='Original Adam')
plt.plot(data2, label='Modified Adam')

# Add labels and title
plt.xlabel('Training Steps')
plt.ylabel('Loss')
plt.title('Training Loss Performance Comparison between Original and Updated Adam')

# Add legend
plt.legend()

# Show the plot
plt.show()
```
